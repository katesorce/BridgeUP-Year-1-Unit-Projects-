```python
#Import libraries

import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
#Open and read CSV for 2002 chloryphyll data.

csvFile = open("/content/drive/MyDrive/BridgeUP STEM/Unit 2: Climate Change/chlorophyll_mgm3_Dec_2002.csv", 'r') 
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```
